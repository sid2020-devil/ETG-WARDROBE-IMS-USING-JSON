### Reading record.json file 

```python
fd = open("record.json","r")
txt = fd.read()
fd.close()
```

### Printing the txt file 

```python
txt
```

### Importing json library and converting into a dictionary 

```python
import json
record  = json.loads(txt)
record
```

### Giving the details to the new item to be added to inventory 

```python
prod_id = input("Enter the id : ")
name = input("Enter the name of the product : ")
mft_date = input("Enter the manufacturing date :")
qty = int(input("Enter the quantity :"))
price = int(input("Enter the price : "))
discount = int(input("Enter the discount : "))
```

### Adding new item to the inventory 

```python
record[prod_id] = {'Name': name , 'Mft. Date': mft_date , 'Qty': qty ,  'Price' : price , 'Discount': discount}
```

### Printing the record with the new item added 

```python
record
```

### Checking if the qty of a given is  same as that of the qty being given as an input 

```python
id = input("Enter the id : ")
qty = int(input("Enter the qty : "))
for id in record:
    if qty != record[id]['Qty']:
        new_qty = record[id]['Qty'] + qty
        break
    else:
        new_qty = record[id]['Qty']
        break
```

### Updating the record with the new qty value to the item 

```python
record[id]['Qty'] = new_qty
```

### Printing the record with the new value of qty for the item 

```python
record
```

### Checking if the price given as input  is same as that of item price in the dictionary 

```python
id = input("Enter the id :")
price = int(input("Enter the price : "))
if id in record:
    if price != record[id]['Price']:
        new_price = record[id]['Price'] + price
    else:
        new_price = record[id]['Price']
else:
    print("Invalid Input")
```

### Updating the price in the dictionary 

```python
record[id]['Price'] = new_price
```

### Printing the dictionary with updated price 

```python
record
```

### Checking if the discount given as input is same as that of item price in the dictionary

```python
id = input("Enter the id :")
disc = int(input("Enter the discount :"))
if id in record:
    if disc != record[id]['Discount']:
        disc_price = record[id]['Discount'] + disc
    else:
        disc_price = record[id]['Discount']
else:
    print("Invalid input")
```

### Updating the value of discount to dictionary 

```python
record[id]['Discount'] = disc_price
```

### Printing the dictionary with the updated value of discount 

```python
record
```

### Updating the inventory 

```python
js = json.dumps(record)
fd = open("record.json","w")
fd.write(js)
fd.close()
```
